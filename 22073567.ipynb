```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import textwrap

# Load the dataset
file_path = "Dataset(3).csv"
data = pd.read_csv(file_path)

# Set the background style
sns.set(rc={'axes.facecolor': 'greenyellow', 'figure.facecolor': 'maroon', 'text.color': 'black'})

# Create plots
fig, axs = plt.subplots(nrows=4, ncols=2, figsize=(28, 32), gridspec_kw={'height_ratios': [0.4, 1, 1, 1.5], 'hspace': 0.9})

title_text = "Global Trends: Population Growth, Unemployment Rates, GDP Growth, and Merchandise Trade Dynamics (2001-2022)"
student_name = 'Umair Ahmad'
student_id = '22073567'

axs[0, 0].remove()  # Remove the existing subplot
axs[0, 1].remove()  # Remove the first row in the second column

axs[0, 0] = fig.add_subplot(4, 2, (1, 2))  

axs[0, 0].wrapped_text = textwrap.fill(title_text, width=55)
axs[0, 0].text(0.5, 0.0, axs[0, 0].wrapped_text, ha='center', va='center', weight='bold', color='white', fontsize=40, transform=axs[0, 0].transAxes)

axs[0, 0].text(0.0, -0.5, student_name, ha='left', va='top', weight='bold', color='white', fontsize=25, transform=axs[0, 0].transAxes)
axs[0, 0].text(0.0, -0.8, student_id, ha='left', va='top', weight='bold', color='white', fontsize=25, transform=axs[0, 0].transAxes)
axs[0, 0].axis('off') 
axs[0, 1].axis('off')  

# Graphs

# Plot 1
# Select specific indicators, countries, and year for analysis
indicators_to_analyze_bar_chart = ['Population growth (annual %)', 'Unemployment, total (% of total labor force) (modeled ILO estimate)', 'GDP growth (annual %)', 'Merchandise trade (% of GDP)']
selected_countries_bar_chart = ['Pakistan', 'India', 'Qatar', 'Japan', 'Australia']
selected_year_first = '2019'

# Loop through each indicator and plot a clustered bar chart
for i, indicator in enumerate(indicators_to_analyze_bar_chart):
    selected_data = data[(data['Indicator Name'] == indicator) & (data['Country Name'].isin(selected_countries_bar_chart))]
    values = selected_data[selected_year_first].astype(float)

    bar_positions = np.arange(len(selected_countries_bar_chart))
    bar_width = 0.05
    bar_gap= 0.03

    axs[1, 0].barh(bar_positions + i * len(selected_countries_bar_chart) * bar_width, values, height=bar_width, label=indicator)

    # Add labels to each bar outside the bars
    for j, value in enumerate(values):
        axs[1, 0].text(value + 0.1, bar_positions[j] + i * len(selected_countries_bar_chart) * bar_width - bar_width / 2, f'{value:.1f}', ha='left', va='center', color='black')

# Customize legend and plot
axs[1, 0].legend()
axs[1, 0].invert_yaxis()
axs[1, 0].set_yticks(bar_positions + (len(indicators_to_analyze_bar_chart) - 1) * len(selected_countries_bar_chart) * bar_width / 2, selected_countries_bar_chart, color='black')
axs[1, 0].set_xlabel('Indicator Value', fontweight='bold', fontsize=15, color='white')
axs[1, 0].set_ylabel('Country', fontweight='bold', fontsize=15, color='white')
axs[1, 0].tick_params(axis='x', colors='white')
axs[1, 0].tick_params(axis='y', colors='white')
axs[1, 0].set_title(f'Clustered Bar Chart for {", ".join(selected_countries_bar_chart)} - ({selected_year_first})', fontweight='bold', fontsize=16, color='white')
#axs[1, 0].grid(False)

# Explanation for Plot 1

long_text_1 = 'The data shows 2019 key metrics for selected countries. Pakistan and India grew 1.61% and 1.03%, respectively, while Japan declined -0.14%. Qatar had 0.1% unemployment, whereas India had 6.51%. All GDP grew, with Qatar at 0.69% and India at 3.87%. Qatar dominated merchandise trade as a proportion of GDP at 57.90%, while Japan was 27.87%.'

axs[1, 0].wrapped_text = textwrap.fill(long_text_1, width=68)
axs[1, 0].text(0.03, -0.42, axs[1, 0].wrapped_text, ha='left', va='center', color='greenyellow', fontsize=18, transform=axs[1, 0].transAxes,)

# Plot 2

# Select the specific year for analysis
selected_year_second = '2015'

# Filter data for the selected indicators, countries, and year
selected_data = data[(data['Indicator Name'].isin(indicators_to_analyze_bar_chart)) & (data['Country Name'].isin(selected_countries_bar_chart))]

# Prepare data for the bar chart
bar_positions = np.arange(len(selected_countries_bar_chart))
bar_width = 0.05
bar_gap = 0.05

for i, indicator in enumerate(indicators_to_analyze_bar_chart):
    values = selected_data[(selected_data['Indicator Name'] == indicator)][selected_year_second].astype(float).values

    axs[1, 1].bar(bar_positions + i * (bar_width + bar_gap), values, width=bar_width, label=indicator)

    for j, value in enumerate(values):
        axs[1, 1].text(bar_positions[j] + i * (bar_width + bar_gap) + bar_width / 2, value + 0.1, f'{value:.2f}', ha='center', va='bottom')

# Adding Xticks
axs[1, 1].set_xlabel('Country', fontweight='bold', fontsize=15, color='white')
axs[1, 1].set_ylabel('Values', fontweight='bold', fontsize=15, color='white')
axs[1, 1].tick_params(axis='x', colors='white')
axs[1, 1].tick_params(axis='y', colors='white')
axs[1, 1].set_title(f'Clustered Bar Chart for {", ".join(selected_countries_bar_chart)} - ({selected_year_second})', fontweight='bold', fontsize=16, color='white')
axs[1, 1].set_xticks(bar_positions + (len(indicators_to_analyze_bar_chart) - 1) * (bar_width + bar_gap) / 2)
axs[1, 1].set_xticklabels(selected_countries_bar_chart)
axs[1, 1].legend()

# Explanation for Plot 2

long_text_2 = 'Selected countries showed distinct tendencies in 2015. Qatar had an 8.65% population growth, 0.17% unemployment, and 4.75% GDP growth, backed by a 68.37% goods trade to GDP ratio. With 7.92% unemployment and 7.99% GDP growth, India showed economic growth. Japan saw a slight population reduction, moderate unemployment of 3.39%, and cautious GDP growth of 1.56%.'

axs[1, 1].wrapped_text = textwrap.fill(long_text_2, width=68)
axs[1, 1].text(0.03, -0.42, axs[1, 1].wrapped_text, ha='left', va='center', color='greenyellow', fontsize=18, transform=axs[1, 1].transAxes)

# Plot 3

# Select specific indicators for analysis
indicators_to_analyze_trend_plot = ['Population growth (annual %)', 'Unemployment, total (% of total labor force) (modeled ILO estimate)', 'GDP growth (annual %)', 'Merchandise trade (% of GDP)']

# Select the specific country for analysis
selected_country_third_plot = 'Pakistan'

# Filter data for selected indicators and country
selected_data_third_plot = data[(data['Indicator Name'].isin(indicators_to_analyze_trend_plot)) & (data['Country Name'] == selected_country_third_plot)]

for indicator in indicators_to_analyze_trend_plot:
    indicator_data = selected_data_third_plot[selected_data_third_plot['Indicator Name'] == indicator]
    axs[2, 0].plot(indicator_data.columns[2:], indicator_data.iloc[0, 2:], marker='o', label=indicator)

axs[2, 0].set_xlabel('Year', fontweight='bold', fontsize=15,color='white')
axs[2, 0].set_ylabel('Value', fontweight='bold', fontsize=15, color='white')
axs[2, 0].tick_params(axis='x', colors='white')
axs[2, 0].tick_params(axis='y', colors='white')
axs[2, 0].set_title(f'Time Trends for {selected_country_third_plot} - Trend of growth in Population, GDP, Merchandise trade and unemployement', fontdict={'fontweight': 'bold', 'color': 'white', 'fontsize' : '12'})
axs[2, 0].legend()

# Explanation for Plot 3

long_text_3 = 'Key data from 2001 to 2022 show constant population increase, starting at 3.09% and dropping to 1.89%. Unemployment averages below 7%, peaking in 2013. GDP growth fluctuates between 7.83% in 2004 and -1.27% in 2012, demonstrating economic resilience. A dynamic economy is shown by fluctuating merchandise trade as a percentage of GDP. '

axs[2, 0].wrapped_text = textwrap.fill(long_text_3, width=68)
axs[2, 0].text(0.03, -0.42, axs[2, 0].wrapped_text, ha='left', va='center', color='greenyellow', fontsize=18, transform=axs[2, 0].transAxes)

# Plot 4

# Select the specific country for analysis
selected_country_fourth_plot = 'Japan'

# Filter data for selected indicators and country
selected_data_fourth_plot = data[(data['Indicator Name'].isin(indicators_to_analyze_trend_plot)) & (data['Country Name'] == selected_country_fourth_plot)]

for indicator in indicators_to_analyze_trend_plot:
    indicator_data = selected_data_fourth_plot[selected_data_fourth_plot['Indicator Name'] == indicator]
    axs[2, 1].plot(indicator_data.columns[2:], indicator_data.iloc[0, 2:], marker='o', label=indicator)

axs[2, 1].set_xlabel('Year', fontweight='bold', fontsize=15,color='white')
axs[2, 1].set_ylabel('Value', fontweight='bold', fontsize=15, color='white')
axs[2, 1].tick_params(axis='x', colors='white')
axs[2, 1].tick_params(axis='y', colors='white')
axs[2, 1].set_title(f'Time Trends for {selected_country_fourth_plot} - Trend of growth in Population, GDP, Merchandise trade and unemployement', fontdict={'fontweight': 'bold', 'color': 'white', 'fontsize' : '12'})
axs[2, 1].legend()
# Explanation for Plot 4

long_text_4 = 'Japan\'s demographic profile shows population growth declining, reaching negative values by 2015. Despite this, unemployment stayed at 2-5%. The economy struggled in 2009 and 2012, but it recovered in succeeding years. Japanese merchandise trade as a proportion of GDP fluctuated between 17.20% and 38.85%, demonstrating its global trading strength. This analysis illuminates Japan\'s economic resiliency and problems across two decades.'

axs[2, 1].wrapped_text = textwrap.fill(long_text_4, width=72)
axs[2, 1].text(0.03, -0.4, axs[2, 1].wrapped_text, ha='left', va='center', color='greenyellow', fontsize=18, transform=axs[2, 1].transAxes)

#plot 5

axs[3, 0].remove()  # Remove the existing subplot
axs[3, 1].remove()  # Remove the fourth row in the second column

axs[3, 0] = fig.add_subplot(4, 1, 4)  

# Select specific indicators for analysis
indicators_to_analyze_cluster_plot = ['Population growth (annual %)', 'Unemployment, total (% of total labor force) (modeled ILO estimate)', 'GDP growth (annual %)', 'Merchandise trade (% of GDP)']

# Select the specific country for analysis
selected_country_first = 'Qatar'

# Select the range of years for analysis
start_year_cluster_plot = 2010
end_year_cluster_plot = 2020

# Filter data for the selected indicators, country, and years
selected_data_first = data[(data['Indicator Name'].isin(indicators_to_analyze_cluster_plot)) & (data['Country Name'] == selected_country_first)]
selected_years_first = [str(year) for year in range(start_year_cluster_plot, end_year_cluster_plot + 1)]

# Prepare data for the bar chart
bar_positions = np.arange(len(selected_years_first))
bar_width = 0.2

for i, indicator in enumerate(indicators_to_analyze_cluster_plot):
    values = selected_data_first[selected_data_first['Indicator Name'] == indicator][selected_years_first].astype(float).values.flatten()

    axs[3, 0].bar(bar_positions + i * bar_width, values, width=bar_width, label=indicator)

# Adding Xticks
axs[3, 0].set_xlabel('Year', fontweight='bold', fontsize=15, color='white')
axs[3, 0].set_ylabel('Values', fontweight='bold', fontsize=15, color='white')
axs[3, 0].set_title(f'Clustered Bar Chart for {selected_country_first}', weight='bold', color='white', fontsize=30)
axs[3, 0].set_xticks(bar_positions)
axs[3, 0].tick_params(axis='x', colors='white')
axs[3, 0].tick_params(axis='y', colors='white')
axs[3, 0].set_xticklabels(selected_years_first)
axs[3, 0].legend()

# Explanation for plot 5

title_text5 = "Qatar's demographic and economic data from 2010 to 2020 show dramatic changes. Population growth peaked at 8.65% in 2015, but a sharp decrease to -1.68% in 2020 signals a reversal. The labor market was stable with unemployment between 0.1 and 0.56%. GDP growth peaked at 19.59% in 2010 and fell to -3.56% in 2020. The economy is strong, with merchandise trade as a percentage of GDP ranging from 53.55% to 87.64%. The heading may be \"Qatar's Economic Dynamics: Population Fluctuations, Stable Employment, and Resilient Trade (2010-2020).\""

axs[3, 0].wrapped_text = textwrap.fill(title_text5, width=145)
axs[3, 0].text(0.5, -0.22, axs[3, 0].wrapped_text, ha='center', va='center', weight='bold', color='greenyellow', fontsize=18, transform=axs[3, 0].transAxes)

 

# Save the figure as a PNG file
plt.show()



```
